# Informe de Métricas y Comparación de Modelos
Carga `models/metrics_full.csv`, ranking por RMSE/MAE y gráficos comparativos por ticker.


```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
m = Path('models/metrics_full.csv')
df = pd.read_csv(m) if m.exists() else pd.DataFrame()
df.head()
```

```python
if not df.empty:
    # ranking por ticker y modelo
    grp = df.groupby(['ticker','model']).agg({'rmse':'mean','mae':'mean','directional_acc':'mean'}).reset_index()
    display(grp.sort_values(['ticker','rmse']).head(20))

    # gráfico RMSE promedio por modelo
    avg = grp.groupby('model')['rmse'].mean().sort_values()
    plt.figure()
    avg.plot(kind='bar')
    plt.ylabel('RMSE promedio')
    plt.title('Comparación global de modelos (RMSE)')
    plt.tight_layout()
    plt.show()
else:
    print('No hay metrics_full.csv aún. Ejecuta el entrenamiento primero.')
```
